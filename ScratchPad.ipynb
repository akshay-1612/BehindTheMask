```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
df = pd.read_csv('./Tweets/mask.csv')
```

```python
df
```

```python
def check_for_img(name):
    for item in df.entities:
        i = ast.literal_eval(item)
        media = i.get('media', [])
        media_file = {item['media_url']:item for item in media}
        for f in media_file:
            if name in f:
                return df['entities'].index.tolist()
```

```python
check_for_img('/1oBJn6Oep2rnO1EH.jpg')
```

```python
for item in df.entities:
    i = ast.literal_eval(item)
    media = i.get('media', [])
    media_file = {item['media_url']:item for item in media}
    for f in media_file:
        if '/1oBJn6Oep2rnO1EH.jpg' in f:
            print(df.item.index.tolist())
```
